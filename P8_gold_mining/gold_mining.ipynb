Визуализация информации по исходным датафреймам показала, что имеются пропуски данных.
Создал функцию для подсчета пропусков и их процента по признакам и в общем,дубликатов и размера.

```python
def mis_inf(data): # функция для подсчета пропусков и их процента по признакам и в общем,дубликатов и размера.
    total = data.isna().sum().sort_values(ascending = False)
    percent = (data.isna().sum()/data.isnull().count()*100).sort_values(ascending = False)
    print(f'Размер датафрейма:',  data.shape)
    print(f'Количество дубликатов датафрейма:', data.duplicated().sum())
    print(f'Количество пропусков датафрейма:',  total.sum())
    print(f'Процент пропусков датафрейма:',  round(percent.sum(),2))
    return pd.concat([total, percent], axis=1, keys=['Total', 'Percent'])
```

```python
print("выборка TRAIN")
mis_inf(train)
```

```python
print("выборка TEST")
mis_inf(test)
```

### Найдем MAE между расчётными и фактическим значением признака. 

```python
c = train['rougher.output.concentrate_au']#доля золота в концентрате после флотации/очистки 
f = train['rougher.input.feed_au'] #доля золота в сырье/концентрате до флотации/очистки
t = train['rougher.output.tail_au'] #доля золота в отвальных хвостах после флотации/очистки
recovery = ((c*(f-t))/(f*(c-t))) * 100 #расчетные значения по формуле

mae=mean_absolute_error(train['rougher.output.recovery'], recovery)

print('MAE -','{0:.20f}'.format(mae))
```

**Ошибка практически нулевая,это значит,что расчеты в датасете и вычисленные по формуле совпадают.**

### Посмотрим, как меняется концентрация металлов (Au, Ag, Pb) на различных этапах: в сырье, в черновом концентрате, в концентрате после первой очистки и в финальном концентрате. 

```python
position = 0
plt.figure(figsize=[15, 17])
plt.subplots_adjust(top=0.92, wspace=0.2, hspace=0.3)
plt.suptitle('ГРАФИК КОНЦЕНТРАЦИИ МЕТАЛЛОВ НА ЭТАПАХ ОЧИТСКИ.', fontsize=15)
for metal in ['au', 'ag', 'pb']:
    position += 1
    sns.set_style('darkgrid')
    plt.subplot(3, 1, position)
    plt.title(f'Распределение {str(metal).upper()} на этапах очистки', fontsize=13)
    #sns.set_style('darkgrid')
    sns.histplot(train[f'rougher.input.feed_{metal}'], color='#E69F00', label='Сырье',kde=True)
    sns.histplot(train[f'rougher.output.concentrate_{metal}'], color='#56B4E9', label='Флотация',kde=True)
    sns.histplot(train[f'primary_cleaner.output.concentrate_{metal}'], color='#F0E442', label='Первичная очистка',kde=True)
    sns.histplot(train[f'final.output.concentrate_{metal}'], color='#009E73', label='Финал',kde=True)
    plt.xlabel('Концентрация металла')
    plt.ylabel('Количество наблюдений')
    plt.legend()
plt.show();
```

Анализ концентрации металлов показал, что концентрация золота значительно увеличивается на каждом этапе обработки сырья, приблизительно в 5 раз. Концентрация Ag уменьшается от этапа к этапу и на финише в среднем в 2 раза меньше чем в сырье. Содержание Pb также увеличивается на финише но стабилизируется и практически не отличается на финише после первичной очистки.На всех графиках распределения концентрации металлов видны нулевые выбросы.

### Исследуйте суммарную концентрацию металлов на разных стадиях: в сырье, в черновом концентрате, в концентрате после первой очистки и в финальном концентрате.

Создал процедуру суммирования концентраций металлов на назных стадиях очистки.

```python
sum_rougher_input = 0
sum_rougher_output = 0
sum_primary_cleaner_output = 0
sum_final_output = 0
def sum_metal(df): # процедура суммирования концентраций металлов на назных стадиях очистки
    df['sum_rougher_input'] = sum_rougher_input
    df['sum_rougher_output'] = sum_rougher_output 
    df['sum_primary_cleaner_output'] = sum_primary_cleaner_output
    df['sum_final_output'] = sum_final_output    
    for metal in ['au', 'ag', 'sol', 'pb']:
            df['sum_rougher_input'] += df[f'rougher.input.feed_{metal}']
            df['sum_rougher_output'] += df[f'rougher.output.concentrate_{metal}']
            df['sum_primary_cleaner_output'] += df[f'primary_cleaner.output.concentrate_{metal}']
            df['sum_final_output'] += df[f'final.output.concentrate_{metal}']
```

### Функция метрики качества

Создал функции для расчета метрик sMAPE  на этапе сырья и на финале, и суммарную SMAPE, по ней и стоит подбирать гиперпараметры и выбирать модель. 

```python
def get_smape(target, predict):#функция расчета SMAPE 
    smape = (1/len(target))*np.sum((np.abs(target - predict))/( (np.abs(target)+np.abs(predict))/2))*100
    return smape
```

```python
def sum_smape(rougher,final): #функция суммарной SMAPE    
    smape_sum=0.25*rougher+0.75*final    
    return smape_sum
```

### Проверка на адекватность

Проверим нашу финальную модель на адекватность, для чего создадим простейшую (dummy) модель, которая всегда предсказывает наиболее часто встречающийся класс. Так мы получим контрольные данные для сравнительной оценки построенной нами более сложной модели. К тому же мы сможем быть уверены в качестве ее работы, например, в том, что она выдает не просто набор случайно подобранных данных.

```python
dummy_rougher = DummyRegressor(strategy="mean")
dummy_rougher.fit(features_train, target_train_rougher)
dummy_predicted_rougher = dummy_rougher.predict(features_test)


dummy_final = DummyRegressor(strategy="mean")
dummy_final.fit(features_train, target_train_final)
dummy_predicted_final = dummy_final.predict(features_test)


dummy_smape_rougher = get_smape(target_test_rougher, dummy_predicted_rougher)
dummy_smape_final = get_smape(target_test_final, dummy_predicted_final)
dummy_smape_total = sum_smape(dummy_smape_rougher, dummy_smape_final)
print("sMAPE_DUMMY_TOTAL:", round(dummy_smape_total,4))
```

Качество модели RandomForestRegressor гораздо выше чем простейшей модели Dummy, это означает, что наша модель вполне вменяема и эффективна.
